```python
df11 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3']
                   },
                    index = [0,1,2,3]
                  )

df12 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                     'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']
                   },
                    index =[0,1,2,3]
                  )
```

```python
df11
```

```python
df12
```

```python
pd.merge(df11,df12,how='inner',on='A')
```

```python
pd.merge(df11,df12, how='outer', on='A') 
```

```python
pd.merge(df11,df12,how='outer',on=['A','C'])
```

```python
df11.join(df12)
```
